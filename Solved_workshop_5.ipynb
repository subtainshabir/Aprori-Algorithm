# Portugal_online_retail

```python
#Import Libraries
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Reading the CSV file 'Portugal_online_retail.csv' and storing it in a DataFrame named 'data'
## Download Dataset from "https://drive.google.com/file/d/1nkQWPu87rrgkqccDYEpzcdXZUL7ua1d8/view?usp=sharing"
data = pd.read_csv('Portugal_online_retail.csv')
```

```python
# Displaying the first few rows of the DataFrame 'data' to inspect its structure and content
data.head()
```

```python
# Dropping the 'InvoiceNo' column from the DataFrame 'data' along the columns axis (axis=1)
data = data.drop(columns='InvoiceNo', axis=1)
```

```python
data.head()
```

```python
## Find Frequent Items
freq_items=apriori(data, min_support=0.1, use_colnames=True)
```

```python
# Sorting the frequent itemsets DataFrame 'freq_items' by the 'support' column in descending order
freq_items.sort_values(by=['support'], ascending=False)
```

```python
# Generating association rules from frequent itemsets with a confidence level of 50%
confidence_rule_1 = association_rules(freq_items, metric='confidence', min_threshold=0.5)

# Sorting the association rules DataFrame 'confidence_rule_1' by confidence in descending order
confidence_rule_1_sorted = confidence_rule_1.sort_values(by='confidence', ascending=False)

# Extracting relevant columns ('antecedents', 'consequents', 'confidence', 'support') from the sorted DataFrame
confidence_rule_df_1 = confidence_rule_1_sorted[['antecedents', 'consequents', 'confidence', 'support']]

# Displaying the DataFrame containing association rules satisfying the confidence threshold
print(confidence_rule_df_1)
```

```python
# Generating association rules from frequent itemsets with a confidence level of 60%
confidence_rule_2 = association_rules(freq_items, metric='confidence', min_threshold=0.6)

# Sorting the association rules DataFrame 'confidence_rule_2' by confidence in descending order
confidence_rule_2_sorted = confidence_rule_2.sort_values(by='confidence', ascending=False)

# Extracting relevant columns ('antecedents', 'consequents', 'confidence', 'support') from the sorted DataFrame
confidence_rule_df_2 = confidence_rule_2_sorted[['antecedents', 'consequents', 'confidence', 'support']]

# Displaying the DataFrame containing association rules satisfying the confidence threshold
print(confidence_rule_df_2)
```

```python
# Confidence Level with 70%
confidence_rule_3=association_rules(freq_items, metric='confidence',min_threshold=0.7)
confidence_rule_3_sorted = confidence_rule_3.sort_values(by='confidence', ascending=False)
confidence_rule_df_3=confidence_rule_3_sorted[['antecedents','consequents','confidence','support']]
print(confidence_rule_df_3)
```

#  Report 

By applying different confidence level I observed the following observations:
* It is oberserved the *lower confidence levels* result in a higher number of association rules being generated, while higher confidence levels produce fewer rules.
* Increasing the confidence level leads to the emergence of *stronger relationships* between items in the rules. This indicates a *tighter association* or simultaneous purchase of items.
* A confidence level of 0.5 is deemed preferable for the Portugal online retail dataset. This choice generates a greater number of rules, providing retailers with more options for organizing items collectively.
* An example association rule with a confidence level of 1 has been highlighted for the Portugal online retail dataset. In this case, the antecedent includes "JUMBO BAG PINK VINTAGE PAISLEY" and "JUMBO BAG RED", while the consequent is "JUMBO SHOPPER VINTAGE RED PAISLEY" with a confidence of 100%. This indicates a strong association where customers purchasing the pink or red vintage paisley jumbo bag also tend to buy the vintage red paisley jumbo shopper.
* Similar trends and observations regarding confidence levels and association rule generation have been observed across all three datasets, including Portugal, Sweden, and the UK online retail markets.
